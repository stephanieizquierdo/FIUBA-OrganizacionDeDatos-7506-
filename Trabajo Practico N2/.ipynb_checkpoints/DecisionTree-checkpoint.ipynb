## Árbol de decisión

Los parámetros que se deben definir son:

- **Profundidad máxima**: le impone un limite a la profundidad de máxima al árbol.

- **Criterio**: es el criterio con el que se escoge el "mejor feature" para ocupar un nodo. Decimos que un feature es mejor que otro si maximiza la ganancia de información o si tiene baja impureza según el criterio de impureza Gini.

- **Cantidad minima de muestras**: es la cantidad minima de muestras que puede tener las hojas del arbol. Esto es para evitar tener hojas con 1 muestra y asi evitar el overfitting

Construiremos cuatro árboles de decision: un primer árbol utilizando un preprocesamineto que convierte las variables categóricas a numéricas (y las numericas que previamente estaban en el dataset se mantienen), un segundo árbol utilizando un preprocesamiento que normaliza todas la columnas. El tercer arbol ademas de tener las variables categoricas a numericas tendra un encodeo numerico y el cuarto ademas los normalizara.

A priori, esperamos que entre el 1,2 y 3,4 árbol sean similares, pues normalizar los valores en un árbol cambia el criterio de corte de sus nodos pero no hay cálculos de distancias, por lo que el árbol no es afectado notablemente por no normalizar.

```python
min_samples_leafs = np.arange(20, 40)

parameters = {'criterion': ['entropy', 'gini'], 'max_depth': np.arange(8, 20),
              'min_samples_leaf': min_samples_leafs}
```

### Version Nro 1

Hacemos one hot encoding con los features categoricos

```python
columns_to_encoder = ["estado_marital", "categoria_de_trabajo", "rol_familiar_registrado", "religion", "trabajo", "genero", "relacion_de_dependencia"]
X_data_1 = encode_one_hot(X_data, columns_to_encoder)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_1.drop(columns={"tiene_alto_valor_adquisitivo"}), X_data_1["tiene_alto_valor_adquisitivo"], random_state=42)
```

```python
decision_tree_1 = DecisionTreeClassifier()
gs_1 = GridSearchCV(decision_tree_1, parameters, n_jobs = -1, scoring='roc_auc')
gs_1.fit(X_train, y_train)
gs_1.best_params_
```

```python
decision_tree_1 = DecisionTreeClassifier( criterion='gini', max_depth=14, min_samples_leaf = 36)
decision_tree_1.fit(X_train, y_train)
y_pred_1 = decision_tree_1.predict(X_test)
```

```python
show_metrics(decision_tree_1, y_pred_1, y_test, X_test, "Arbol 1")
```

Para esta primer version para el modelo de arbol de decision observamos que tenemos un buen accuracy y un recall que esta bien, pero vemos que el precision es bajo con respecto a las otras metricas, vamos a ver si con las siguientes versiones esto mejora.

```python
show_roc_curve(decision_tree_1,"Arbol 1",X_test, X_train, y_test, y_train)
```

```python
show_confusion_matrix(y_pred_1,y_test)
```
